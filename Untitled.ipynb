```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from time import sleep
from selenium.webdriver.chrome.options import Options
import sys
import regex as re
import pandas as pd
```

```python
nombres = pd.read_csv('Data/name.csv')
Metros = pd.read_csv('Data/metros.csv')
pistas = pd.read_csv('Data/pistas.csv')
remontes = pd.read_csv('Data/remontes.csv')
precios = pd.read_csv('Data/precios.csv')
temporada = pd.read_csv('Data/Season.csv')
```

```python
dfINAL = pd.concat([nombres,Metros,pistas,remontes,precios,temporada], axis=1)
```

```python
dfINAL = dfINAL.drop('Unnamed: 0', 1)
```

```python
dfINAL
```

```python
def limpia_col(dfcol):
    for i in dfcol:
        a = dfcol.str.extract('((\d+)+(\.\d+)|(\d+))', flags=0)
        a = a.rename(columns={0:"Primera"})
    return a.Primera
```

```python
dfINAL['Minima Altura'] = limpia_col(dfINAL['Minima Altura'])
dfINAL['Maxima Altura'] = limpia_col(dfINAL['Maxima Altura'])
dfINAL['Adults'] = limpia_col(dfINAL['Adults'])
dfINAL['Children'] = limpia_col(dfINAL['Children'])
dfINAL['Youth'] = limpia_col(dfINAL['Youth'])
```
