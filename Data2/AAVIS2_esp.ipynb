```python
from geopy.geocoders import Nominatim 
```

```python
def coger_loc(columnaNombre):
    try:
        geolocator = Nominatim(user_agent="Jaime")
        location = geolocator.geocode(f'{columnaNombre}')
        x = location.address

        return list(( x.split(",")[-3], (location.latitude, location.longitude)))
    except:
        return 'error'
    
    
    
```
